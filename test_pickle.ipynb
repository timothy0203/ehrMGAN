```python
import pickle
import numpy as np

# Step 1: Load the .pkl file
with open('data/real/mimic/vital_sign_24hrs_p4.pkl', 'rb') as f:
    data = pickle.load(f)

# Step 2: Save the data to a .txt file
np.savetxt('data/real/mimic/vital_sign_24hrs.txt', data.reshape(-1, data.shape[-1]))

# Step 3: Read the data back from the .txt file
loaded_data = np.loadtxt('data/real/mimic/vital_sign_24hrs.txt')

# Step 4: Reshape the data to the original shape
loaded_data = loaded_data.reshape(28344, 24, 104)

# Verify the shape
print(loaded_data.shape)
```

```python
import pickle
import numpy as np
import pandas as pd

# Step 1: Load the .pkl file
with open('data/real/mimic/statics.pkl', 'rb') as f:
    data = pickle.load(f)

# Convert DataFrame to NumPy array, ensuring all data is numeric
data_array = data.apply(pd.to_numeric, errors='coerce').values

# Step 2: Save the data to a .txt file
np.savetxt('data/real/mimic/statics.txt', data_array, fmt='%.18e')

# Step 3: Read the data back from the .txt file
loaded_data = np.loadtxt('data/real/mimic/statics.txt')

# Step 4: Reshape the data to the original shape
loaded_data = loaded_data.reshape(28344, 28)

# Verify the shape
print(loaded_data.shape)
```
